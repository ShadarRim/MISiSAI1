## Списки

Вам дан список. Поменяйте порядок следования элементов, начиная с середины до конца списка.

```python
import numpy as np

a = [6,5,2,4,8,1,0,2]

a[len(a)//2:] = sorted(a[len(a)//2:])
a
```

Вам дан список. Поменяйте в нём местами блоки элементов: каждый чётный со второго по пятый с каждым нечётым со второго по пятый с конца. 

```python
a = [6,5,2,4,8,1,0,2,0,9]
x1 = a[-5:-1]
x1[::2] = a[1:5][::2]

y2 = a[1:5]
y2[::2] = a[-5:-1][::2]
a[1:5] = y2
a[-5:-1] = x1
```

```python
a = [6,5,2,4,8,1,0,2,0,9]

a[1:5:2],a[-5:-1:2] = a[-5:-1:2],a[1:5:2]
a
```

## Генераторы

Вам дан список. Создайте новый список, который возведёт в квадрат элементы исходного списка.

```python
a = [6,5,2,4,8,1,0,2,0,9]

a2 = [i**2 for i in a]
a2
```

## NumPy

Вам дан одномерный массив. На основе этого массива сгенерируйте двумерный массив, который группирует по 4 элемента в строку из исходного.

Размерность исходного массива кратна 4.

Например:

`A = [1,2,3,4,5]` - исходный массив

`B = [[1,2,3,4],[2,3,4,5]]` - полученный

```python
import numpy as np
a = [1,2,3,4,5]
#np.split(a,4)

b = [a[i:i+4] for i in range(0,len(a),4)]
b[-1] = a[-4:]
b
```

```python
db['Ram'] = db['Ram'].apply(lambda x: int(x[:-2]))
```

```python
db.insert(6,'Height',[int(i.split('x')[-1]) for i in db['ScreenResolution'].tolist()])
db = db.rename(columns = { 'ScreenResolution' : 'Width'})
db['Width'] = [int(i.split('x')[0].split(' ')[-1]) for i in db['Width'].tolist()]
```

```python
#[i.split(' ')[-1] for i in db['Cpu'].tolist()]
db.insert(8,'Frequency',[float(i.split(' ')[-1].split('GHz')[0]) for i in db['Cpu'].tolist()])
db['Cpu'] = [i[:i.rfind(' ')] for i in db['Cpu'].tolist()]
```

```python
db['Weight'] = [float(i.split('kg')[0]) for i in db['Weight'].tolist()]
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10,10))
sns.heatmap(db.corr(),xticklabels = db.corr().columns,yticklabels = db.corr().columns, center=0,annot = True)
```

```python
sns.pairplot(db, kind="reg")
plt.show()
```

```python
fig, axes = plt.subplots(1, 3, figsize=(15,5)) 
sns.boxplot(db['Ram'],ax = axes[0])
sns.boxplot(db['Price_euros'],ax = axes[1])
sns.boxplot(db['Frequency'],ax = axes[2])
```

```python
Q1 = db['Ram'].quantile(0.25)
Q3 = db['Ram'].quantile(0.75)
IQR = Q3 - Q1
mask = db['Ram'] > Q3+1.5*IQR
db[mask]
```

```python
Q1 = db['Price_euros'].quantile(0.25)
Q3 = db['Price_euros'].quantile(0.75)
IQR = Q3 - Q1
mask = (db['Price_euros'] > Q3+1.5*IQR) | (db['Price_euros'] < Q1-1.5*IQR)
db[mask]
```

```python
Q1 = db['Frequency'].quantile(0.25)
Q3 = db['Frequency'].quantile(0.75)
IQR = Q3 - Q1
mask = db['Frequency'] < Q1-1.5*IQR
db[mask]
```
