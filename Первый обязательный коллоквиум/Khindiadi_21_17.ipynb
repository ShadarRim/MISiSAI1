## Списки

Вам дан список. Создайте новый список, в котором порядок следования элементов будет обратным относительно исходного списка. 

```python
s = [int(i) for i in input().split()]
res = s[::-1]
res
```

Вам дан список. Создайте новый список, в котором будут присутствовать только чётные элементы исходного списка.

```python
s = [int(i) for i in input().split()]
res = []
for i in s:
  if i%2==0:
    res.append(i)
res
```

```python
s = [int(i) for i in input().split()]
s[::2]
```

## Генераторы

Вам дан список. Создайте новый список, который удвоит каждый элемент исходного списка.

```python
s = [int(i) for i in input().split()]
res = [i*2 for i in s]
res
```

Вам дан список. Создайте новый список, который умножит увеличит на треть элемент исходного списка.

```python
s = [int(i) for i in input().split()]
res = [i*1/3 for i in s]
res
```

```python
s = input()
s1 = [ord(i) for i in s]
s2 = [i for i in range(len(s))]
tuple(zip(s2,s1))
```

## NumPy

Создайте многомерный массив, посчитайте суммы элементов по различным осям.

```python
import numpy as np
n, m = [int(i) for i in input('Введите размерность массива через пробел: ').split()]
s = np.random.randint(1,10,size = (n,m))

ax = int(input("Введите:   0 - по вертикали,   1 - по горизонтали    "))
print(np.sum(s,axis = ax))
```

Дан массив A (numpy.array()), заполненный 1.
Дан массив I (numpy.array()) индексов.

Прибавьте к каждому значению массиву A по массиву индексов +1.

```python
import numpy as np

n = 10
k = 12
A = np.ones(n)
I = np.random.randint(0,len(A),k)
for i in I:
  A[i] +=1 
print(A)
```

- Постройте boxplot для всех переменных и выведете объекты которые являются выбросами (строки из датафрейма).
- Посчитайте различные статестические показатели (среднее, медиана, максимум, минимум) для различных классов *(т.е. отдельно какие показатели у класса 0, какие у класса 1)*.
- Проанализируйте колонку `SpType`. Подсчитаете количество звёзд относящихся к различным классам.
`SpType` - спектральный класс звезды согласно классификации Моргана-Кинана.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('Star3642_balanced.csv')

#1 Количество типов данных
print(f"1. Количество типов данных = {len(df.dtypes.unique())}")
print(df.dtypes.unique())
#2 Кол-во пропущенных значений
n = df.isnull().sum().sum()
if n:
  print(f"2. Да, таких значений: {n}")
else:
  print("2. Нет")
```

```python
#3 Гистограмма Target
df['TargetClass'].plot(kind = 'hist', figsize = (10,10), title = "Hist of Target")
```

```python
#4 Heatmap
print('4. Heatmap')
plt.figure(figsize=(8,8), dpi= 80)
sns.heatmap(df.corr(), xticklabels=df.corr().columns, yticklabels=df.corr().columns, cmap='RdYlGn', center=0, annot=True)

plt.title('Heatmap of Star3642', fontsize=22)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.show()
```

```python
#5 Pairplot 
print('5. Pairplot of Star3642')
sns.pairplot(df,size = 1)
plt.show()
```

```python
#6 Boxplot
print('6. Boxplot')
df.plot(kind='box', figsize = (8,8), title = 'Boxplot of Star3642')
```

```python
#7 Показатели
for i in df.keys():
  if i != "SpType":
    print(f"Класс {i}")
    print(f"Медиана для класса {i} = {df[i].median()}")
    print(f"Среднее для класса {i} = {df[i].mean()}")
    print(f"Максимальное для класса {i} = {df[i].max()}")
    print(f"Минимальное для класса {i} = {df[i].min()}")
    print()
```

```python
#8
k = {} 
for i in df['SpType']:
  if i[0] in k:
    k[i[0]]+=1
  else:
    k[i[0]] = 1
for i in k:
  print(f"Звезд класса {i}: {k[i]}")
```
