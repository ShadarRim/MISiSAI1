## Списки

Вам дан список. Поменяйте порядок следования элементов, начиная с середины до конца списка.

```python
# ваше решение
a = [1, 2, 3, 4, 5, 6, 7, 8, 9]
a = a[:len(a) // 2] + a[-1:len(a) // 2 - 1:-1]
a
```

Вам дан список. Поменяйте в нём местами блоки элементов: каждый чётный со второго по пятый с каждым нечётым со второго по пятый с конца. 

```python
# ваше решение
a = [1, 2, 3, 4, 5, 6, 6, 6, 6]
chet = [i for i in range(1, 5) if a[i]%2 == 0]
nechet  = [i for i in range(len(a) - 5, len(a)) if a[i]%2 == 1]
for i in range(min(len(chet), len(nechet))):
    a[chet[i]], a[nechet[i]] = a[nechet[i]], a[chet[i]]
a
```

```python
a = [1, 2, 3, 4, 5, 6, 6, 6, 6]
a[1:5:2], a[-5:-1:2] =  a[-5:-1:2], a[1:5:2]
a
```

## Встроенные функции

Вам дана строка. Выведите разницу между наибольшим и наименьшим ascii-кодом символа в строке. 

```python
# ваше решение
s = 'azslwd;cwlvwlv'
maxx, minn = ord(s[0]), ord(s[0])
for val in s:
    if maxx < ord(val):
        maxx = ord(val)
    if minn > ord(val):
        minn = ord(val)
maxx - minn
```

```python
import pandas as pd

df = pd.read_csv('laptop_price.csv', encoding='cp1251')
df.head()
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
df['Ram'] = df['Ram'].map(lambda x : int(x[:-2]))
df['Weight'] = df['Weight'].map(lambda x : float(x[:-2]))
df['LenthScreen'] = df['ScreenResolution'].map(lambda x : int(x.split()[-1].split('x')[1]))
df['WidthScreen'] = df['ScreenResolution'].map(lambda x : int(x.split()[-1].split('x')[0]))
df['TypeCpu'] = df['Cpu'].map(lambda x : ' '.join(x.split()[:-1]))
df['Cpu_GHz'] = df['Cpu'].map(lambda x : float(x.split()[-1][:-3]))
df['TypeMemory'] = df['Memory'].map(lambda x : ' '.join(x.split()[1:]))
df['Memory'] = df['Memory'].map(lambda x : float(x.split()[0][:-2]) * 1000 if x.split()[0][-2:] == 'TB' else float(x.split()[0][:-2]))
df.head()
```

```python
int_columns = []
for col in df.columns:
    try:
        pd.to_numeric(df[col])
        int_columns.append(col)
    except:
        pass
```

```python
import seaborn as sns

sns.set(rc = {'figure.figsize':(10,10)})
sns.heatmap(df.corr(), annot=True);
```

```python
sns.pairplot(df)
```

```python
sns.boxplot(df)
```

```python
import numpy as np

df_vibros = pd.DataFrame()
for col in int_columns:
    q75 = df[col].quantile(0.75)
    q25 = df[col].quantile(0.25)
    iqr = q75 - q25
    df_vibros = pd.concat([df_vibros, df[df[col] < q25 - 1.5 * iqr]])
    df_vibros = pd.concat([df_vibros, df[df[col] > q75 + 1.5 * iqr]])

df_vibros = df_vibros.drop_duplicates()
df_vibros
```
