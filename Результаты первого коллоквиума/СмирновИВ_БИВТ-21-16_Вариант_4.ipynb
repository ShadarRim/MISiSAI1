## Генераторы

Вам дан список. Создайте новый список, в котором останутся только чётные элементы кратные трём. 

```python
# ваше решение
arr = [0, 3, 6, 9, 11, 15, 18]
arr = [arr[i] for i in range(len(arr)) if arr[i] % 3 == 0 and i % 2 == 0]
arr
```

```python
# ваше решение
arr = [i for i in range(15)]
new_arr = [(arr[i], arr[-i-1]) for i in range(len(arr))]
new_arr
```

## NumPy

Дан массив 16х16
Создайте новый массив 4х4 , элементы которого будут суммой блоков 4х4 исходного массива

```python
# ваше решение
import numpy as np
arr = np.array([[i+j for j in range(16)] for i in range(16)])
print(arr)
new_arr = np.array([[arr[i*4:i*4+4, j*4:j*4+4].sum() for j in range(arr.shape[1]//4)] for i in range(arr.shape[0]//4)])
new_arr
```

Дан массив, вставьте между каждым значением сумму соседних элементов

```python
# ваше решение
arr = np.array([i for i in range(1, 15, 2)])
print(arr)
new_arr = np.insert(arr, np.arange(1, len(arr)), arr[:-1] + arr[1:])
new_arr
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("garments_worker_productivity.csv", sep=',')
```

```python
plt.figure(figsize=(5,5))

sns.distplot(df['actual_productivity'], bins=30)
```

```python
plt.figure(figsize=(12,10), dpi= 80)
sns.heatmap(df.corr(), xticklabels=df.corr().columns, yticklabels=df.corr().columns, cmap='RdYlGn', center=0, annot=True)

plt.title('Correlogram', fontsize=22)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.show()
```

```python
table = df.corr()
columns = table.columns
for col in table:
  i=0
  while columns[i] != col:
    if table[col][i] >= 0.3:
      print((col, columns[i]), 'corr = ', table[col][i])
    i += 1
```

```python
plt.figure(figsize=(10,8), dpi= 80)
sns.pairplot(df)
plt.show()
```

```python
fig, axes = plt.subplots(1, 11, figsize=(30,10))  

i = 0
for col in df.columns:
  if df[col].dtype.name in ('int64', 'float64'):
    boxplot = sns.boxplot(df[col], ax=axes[i], width=0.3)
    boxplot.set_xlabel(col)
    i += 1
```

```python
for col in df.columns:
  if df[col].dtype.name in ('int64', 'float64'):
    Q1_wip = df[col].quantile(0.25)
    Q3_wip = df[col].quantile(0.75)
    IQR_wip = Q3_wip - Q1_wip
    print(f'Выбросы столбца {col}')
    print(df[(df[col] <= Q1_wip - 1.5 * IQR_wip) | (df[col] >= Q3_wip + 1.5 * IQR_wip)])
```

```python
stat_df = df.groupby('quarter').agg(['mean', 'median', 'max', 'min'])
stat_df
```

```python
date_df = df['date'].str.split('/', expand=True)
date_df.columns = ['day', 'month', 'year']
new_df = pd.concat([date_df, df], axis=1)
new_df.drop('date', axis=1)
new_df
```
