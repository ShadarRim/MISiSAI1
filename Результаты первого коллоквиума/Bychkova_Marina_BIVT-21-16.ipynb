## Списки

Вам дан список. Поменяйте местами первую и вторую половину списка.

```python
# ваше решение
a = [i for i in range(1, 7)]
print(a)
a[:len(a)//2], a[len(a)//2:] = a[len(a)//2:], a[:len(a)//2]
a
```

Вам дан список. Поменяйте в нём местами блоки элементов: каждый нечётный со первого по пятый с начала с каждым чётный со второго по пятый с конца. 

```python
# ваше решение
a = [i for i in range(1, 21)]
print(a)
a[1:6:2], a[-2:-7:-2] = a[-2:-7:-2], a[1:6:2]
a
```

## Генераторы

Вам дан список. Создайте новый список, в котором останутся только чётные элементы кратные трём. 

```python
a = [i for i in range(1, 21)]
new_a = [el for el in a if (el % 2 ==0 and el % 3 ==0)]
new_a
```

```python
# ваше решение
a = [i for i in range(1, 11)]
a_new = [(a[i], a[-i-1]) for i in range(len(a))]
a_new
```

## NumPy

Дан массив 16х16
Создайте новый массив 4х4 , элементы которого будут суммой блоков 4х4 исходного массива

```python
# ваше решение
import numpy as np
a = np.arange(256).reshape((16, 16))
print(a)
a_new = [[a[4*i:4*(i+1), 4*j:4*(j+1)].sum() for j in range(4)] for i in range(4)]
a_new
```

Дан массив, вставьте между каждым значением сумму соседних элементов

```python
# ваше решение
import numpy as np
a = np.arange(10)
sum_a = [a[i]+a[i+1] for i in range(0, len(a)-1)]
for i in range(0, len(a)-1):
    a = np.insert(a, i*2+1, sum_a[i])
a
```

- Постройте boxplot для всех переменных и выведете объекты которые являются выбросами (строки из датафрейма).

```python
# plt.figure(figsize=(20, 20))
import seaborn as sns
fig, axes = plt.subplots(1, 11, figsize=(20, 20))
i = 0
for c in df.select_dtypes(include=['number']):
    sns.boxplot(df[c], ax=axes[i], width=0.7)
    axes[i].set_title(str(c))
    i += 1

plt.show()
```

```python
import numpy as np
def ul(col):
    q1 = np.quantile(col, 0.25)
    q3 = np.quantile(col, 0.75)
    iqr = q3 - q1
    lower_b = q1 - 1.5*iqr
    upper_b = q3 + 1.5*iqr
    return (lower_b, upper_b)

for c in df.select_dtypes(include=['number']):
    print(f'Выбросы по параметру {c}')
    col = df[c]
    print(col[(col < ul(col)[0]) | (col > ul(col)[1])])
```
