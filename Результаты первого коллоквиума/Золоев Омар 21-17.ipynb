```python
column = 'Plx'
Q1 = data[column].quantile(0.25)
Q3 = data[column].quantile(0.75)
Q2 = Q3 - Q1
outliers = data[(data[column] < (Q1 - 1.5 * Q2)) | (data[column] > (Q3 + 1.5 * Q2))]
print(f"Выбросы в столбце{column}:")
print(outliers), len(data)
```
