Даны линии и точка

Расчитайте расстояние от точки до каждой прямой, найдите ближащую.

```python
import numpy as np

point = np.array([2, 0])

line1 = np.array([[0, 0], [4, -4]])
line2 = np.array([[1, 1], [-1, 5]])
line3 = np.array([[3, 10], [3, 4]])

distances = []
for line in [line1, line2, line3]:
    x1, y1 = line[0]
    x2, y2 = line[1]
    distance = abs(np.cross(np.array([x2 - x1, y2 - y1]), np.array([x1 - point[0], y1 - point[1]]))) / np.linalg.norm(np.array([x2 - x1, y2 - y1]))
    distances.append(distance)

closest_line_index = np.argmin(distances)
closest_line = [line1, line2, line3][closest_line_index]

print(f"Dist from {point} to lines: {distances}")
print(f"Closest line is \n {closest_line}\n ({min(distances)})")
```

```python
import pandas as pd

f = pd.read_csv('Train.csv', encoding='latin-1')
f.dtypes.unique()
```

```python
p = f['Unit'].mode()
med = f[f['Unit'] == p[0]]
meann = med['Attrition_rate'].mean()

med[(med['Relationship_Status'] == 'Married') & (med['Attrition_rate'] > meann)]['Gender'].count()
```
