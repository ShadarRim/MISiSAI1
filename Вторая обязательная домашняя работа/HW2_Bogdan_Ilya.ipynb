```python
import pandas as pd

df = pd.read_csv(r"/content/life-expectancy-prediction/Life_train.csv", sep=',')

df.head(5)
```

```python
df.info()
```

Только численные значения, работы со строками не будет.

```python
df_nan = df.loc[:, df.isnull().any()]
df_nan[df_nan.isnull().any(axis=1)]
```

Пропуски разбросаны по всей таблице, полностью нулевых строк нет. Посмотрим процентное содержание пропусков в столбцах.

```python
df_nan.isnull().sum() / len(df_nan) * 100
```

Процентные содержание пропусков в столбцах у всех небольшое, никакие толбцы удалять не будем.

В зависимости от вида распределения выборок, определимся, каким способом заполним пропуски.

```python
#чтобы не испортить статистику, заполняем средним
df[" BMI "].fillna(df[" BMI "].mean(), inplace=True)
#есть явно выделяющееся популярное значение
df["Polio"].fillna(df["Polio"].mode()[0], inplace=True)
df["Alcohol"].fillna(df["Alcohol"].mode()[0], inplace=True)
df["Hepatitis B"].fillna(df["Hepatitis B"].mode()[0], inplace=True)
df["Diphtheria "].fillna(df["Diphtheria "].mode()[0], inplace=True)
df["GDP"].fillna(df["GDP"].mode()[0], inplace=True)
df["Population"].fillna(df["Population"].mode()[0], inplace=True)
df[" thinness  1-19 years"].fillna(df[" thinness  1-19 years"].mode()[0], inplace=True)
df[" thinness 5-9 years"].fillna(df[" thinness 5-9 years"].mode()[0], inplace=True)
#похоже на нормальное распредление
df["Total expenditure"].interpolate(method="linear", direction = "forward", inplace=True)
df["Income composition of resources"].interpolate(method="linear", direction = "forward", inplace=True)
df["Schooling"].interpolate(method="linear", direction = "forward", inplace=True)
```

```python
df = df.select_dtypes(include=np.number)
df.describe()
```

Видно, что признаки лежат в разных диапазонах, проведем масштабирование данных. Все признаки не меньше нуля, пусть так и останется, воспользуемся нормализацией.

```python
from sklearn.preprocessing import MinMaxScaler
X = df.drop(columns=["Year", "Life expectancy "])
y = df["Life expectancy "]
#не хочется делать значения нулевыми, использую нормализацию
mmsc = MinMaxScaler()
X_norm = mmsc.fit_transform(X)
X_norm = pd.DataFrame(X_norm, columns=X.columns)
X_norm.describe()
```

Проведем аналогичные операции для Тестовой выборки

```python
df_test = pd.read_csv(r"/content/life-expectancy-prediction/Life_test.csv", sep=',')

df_test[" BMI "].fillna(df_test[" BMI "].mean(), inplace=True)
df_test["Polio"].fillna(df_test["Polio"].mode()[0], inplace=True)
df_test["Alcohol"].fillna(df_test["Alcohol"].mode()[0], inplace=True)
df_test["Hepatitis B"].fillna(df_test["Hepatitis B"].mode()[0], inplace=True)
df_test["Diphtheria "].fillna(df_test["Diphtheria "].mode()[0], inplace=True)
df_test["GDP"].fillna(df_test["GDP"].mode()[0], inplace=True)
df_test["Population"].fillna(df_test["Population"].mode()[0], inplace=True)
df_test[" thinness  1-19 years"].fillna(df_test[" thinness  1-19 years"].mode()[0], inplace=True)
df_test[" thinness 5-9 years"].fillna(df_test[" thinness 5-9 years"].mode()[0], inplace=True)
df_test["Total expenditure"].interpolate(method="linear", direction = "forward", inplace=True)
df_test["Income composition of resources"].interpolate(method="linear", direction = "forward", inplace=True)
df_test["Schooling"].interpolate(method="linear", direction = "forward", inplace=True)

X2 = df_test.drop(columns=["Year"])
X_norm2 = mmsc.fit_transform(X2)
X_norm2 = pd.DataFrame(X_norm2, columns=X2.columns)

X_train = X_norm
X_test = X_norm2
y_train = y
```

Обучим модель

```python
from sklearn.linear_model import LinearRegression
from sklearn import metrics 

model = LinearRegression()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
```

Полученные коэффициенты

```python
coeff_df = pd.DataFrame({"Index": np.arange(0, len(y_pred)), "Life expectancy": y_pred})
coeff_df
```

```python
coeff_df.to_csv(r"/content/Predictions.csv", index=False)
```
